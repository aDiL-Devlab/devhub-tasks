```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset("iris")
```

```python
print(f"shape of iris dataset: {iris.shape}\n")
print(f"columns of iris dataset: {iris.columns}\n")
print(f"dtypes of iris dataset: {iris.dtypes}\n")
print(f"missing values in iris dataset: {iris.isnull().sum()}\n")
print(f"dupicate values in iris dataset: {iris.duplicated().sum()}\n")
print(f"top 5 row of iris dataset: {iris.head()}\n")
```

```python
iris = iris.drop_duplicates()
iris.shape
```

#### Observation: The dataset has 150 rows and 5 columns 4 numerical and 1 categorical (species). No missing values found. One duplicate removed, leaving 149 rows. ####

### HEATMAP: ###

```python
# Drawing a heatmap to find high correlation between pairs of variables for scatter plots
iris_correlation = iris.corr(numeric_only=True)
sns.heatmap(iris_correlation, annot=True, cmap="coolwarm")
plt.title("correlation between each variable")
plt.show()
```

#### Observation: Petal length and petal width have the highest correlation (0.96), making them the best candidates for scatter plot analysis. ####

### SCATTER PLOT: ###

```python
#scatter plot to show relationship with petal length and it's width
plt.figure(figsize=(6,4))
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue="species")
plt.title("Scatter Plot: Petal Length vs Petal Width")
plt.show()
```

#### Observation: A strong positive correlation exists between petal length and width. Setosa is clearly separated from other species, while Versicolor and Virginica show slight overlap.

### HISTOGRAM: ###

```python
# Histogram showing the distribution of sepal lengths across different iris species using stacked bars
plt.figure(figsize=(6,4))
sns.histplot(data=iris, x="sepal_length", hue="species", multiple='stack', bins=10, palette="Set1")
plt.title("Distribution of Sepal Length by Species")
plt.show()
```

#### Observstion: Setosa dominates the lower sepal length range, Versicolor the medium, and Virginica the higher range. Setosa is clearly separated while Versicolor and Virginica overlap. ####

```python
# Boxplot to visualize the distribution, medians, and outliers of all numeric Iris features
plt.figure(figsize=(6,4))
sns.boxplot(data=iris.drop(columns=["species"]), palette="muted")
plt.title("Boxplot of Iris Features")
plt.show()
```

#### Observation: Sepal width contains visible outliers. Petal length shows the most variation, making it the most informative feature for species classification. ###

## Conclusion/Summary:
 The Iris dataset contains 4 numerical features sepal length, sepal width, petal length, and petal width and 1 categorical column (species) with 3 classes: Setosa, Versicolor, and Virginica.
Through visualization, we observed a strong positive correlation (0.96) between petal length and petal width as petal length increases, petal width increases as well.
In terms of spread, sepal length has the largest range but petal length shows the most variation across species. Petal width has the smallest range overall, while sepal width stays in the mid range with low variation.
Species-wise, Setosa is unique in having the largest sepal width but the smallest petal dimensions. Virginica has the largest overall size except for sepal width. Versicolor falls in the medium range across all features.
Overall, petal features are the most useful for distinguishing between species, as they show the most variation and least overlap.
